```python
import requests
from bs4 import BeautifulSoup
```

```python
good_books = list()
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
i = 0
while True:
    i+=1
    res = requests.get(base_url.format(i))
    if res.status_code == 200:
        soup = BeautifulSoup(res.text, 'html.parser')
        book_details = soup.select('.product_pod')
        for item in book_details:
            rating_above4 = len(item.select('.star-rating.Four'))+len(item.select('.star-rating.Five'))
            if rating_above4:    
                title = item.select('a')[1].get('title')
                price = item.select('.price_color')[0].text[1:]
                availability = item.select('.instock.availability')[0].text.replace(' ', '').replace('\n', '')
                good_books.append({'title': title, 'price': price, 'availability': availability})
    else:        
        break
print(good_books)        
```

```python
base_link = 'http://quotes.toscrape.com/page/{}/'
i = 0
quotes = list()
while True:
    i+=1   
    response = requests.get(base_link.format(i))
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        quote_details = soup.select('.quote')
        if len(quote_details):
            for item in quote_details:
                quote = item.select('.text')[0].text
                author = item.select('.author')[0].text
                tag_details = item.select('.tag')
                tags = set()
                for tag in tag_details:
                    tags.add(tag.text.capitalize())
                quotes.append({'quote': quote, 'author': author, 'tags': tags})
        else:
            break
    else:
        break
print(quotes)
```
