```python
suits = {'Hearts', 'Spades', 'Clubs', 'Diamonds'}
rank = {'Ace', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King'}
values = {'Ace': 14, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5, 'Six': 6, 'Seven': 7, 'Eight': 8, 'Nine': 9, 'Ten': 10, 'Jack': 11, 'Queen': 12, 'King': 13}

from random import shuffle
```

```python
class Card():
    def __init__(self, suit, rank):
        if suit.capitalize() in suits and rank.capitalize() in rank:
            self.suit = suit.capitalize()
            self.rank = rank.capitalize()
            self.value = values[rank.capitalize()]
        
    def __str__(self):
        return f'{self.rank} of {self.suit} --> value {self.value}'
```

```python
class Deck():
    def __init__(self):
        self.all_cards = [ Card(s, r) for s in suits for r in rank ]
    
    def shuffle(self):
        shuffle(self.all_cards)
    
    def draw(self):
        return self.all_cards.pop()
```

```python
class Player():
    def __init__(self, name):
        self.name = name
        self.all_cards = list()
    
    def __str__(self):
        return self.name + " has " + len(self.all_cards) + " cards"
    
    def draw_card (self):
        return self.all_cards.pop(0)
    
    def add_cards(self, new_card):
        if type(new_card) == type(self.all_cards):
            self.all_cards.extend(new_card)
        else:
            self.all_cards.append(new_card)
```

```python
player1 = Player('Player1')
player2 = Player('Player2')
new_deck = Deck()
new_deck.shuffle()
half_len = int(len(new_deck.all_cards)/2)
# cards_split1 = new_deck.all_cards[:half_len]
# cards_split2 = new_deck.all_cards[half_len:]
# player1.add_cards(cards_split1[::-1])
# player2.add_cards(cards_split2[::-1])
for i in range(half_len):
    player1.add_cards(new_deck.draw())
    player2.add_cards(new_deck.draw())
```

```python
player1_tbl = list()
player2_tbl = list()   
rounds = 0
game_on = True
while game_on:
    rounds +=1
    if len(player1.all_cards) == 0:
        print(f'{player1.name} Lost, {player2.name} Won, Game Over')
        game_on = False
        break
    elif len(player2.all_cards) == 0:
        print(f'{player2.name} Lost, {player1.name} Won, Game Over')
        game_on = False
        break
    else:        
        player1_draw = player1.draw_card()
        player2_draw = player2.draw_card()
        player1_tbl.append(player1_draw)
        player2_tbl.append(player2_draw)
        if player1_draw.value == player2_draw.value:            
            if len(player1.all_cards) < 5:
                print(f'Not Enough Cards for War, {player1.name} Lost, {player2.name} Won, Game Over')
                game_on = False
                break
            elif len(player2.all_cards) < 5:
                print(f'Not Enough Cards for War, {player2.name} Lost, {player1.name} Won, Game Over')
                game_on = False
                break
            else:
                print (f'Round {rounds} --> At War!!!')
            at_war = True
        elif player1_draw.value > player2_draw.value:
            player1_tbl.extend(player2_tbl)            
            print (f'Player 1 wins Round {rounds}, Gets {len(player1_tbl)} Cards!!!')
            player1.add_cards(player1_tbl)
            player1_tbl.clear()
            player2_tbl.clear()
        elif player2_draw.value > player1_draw.value:            
            player2_tbl.extend(player1_tbl)
            print (f'Player 2 wins Round {rounds}, Gets {len(player2_tbl)} Cards!!!')
            player2.add_cards(player2_tbl)
            player1_tbl.clear()
            player2_tbl.clear()
    i = 0
    while at_war and game_on:
        i+=1
        player1_draw = player1.draw_card()
        player2_draw = player2.draw_card()
        player1_tbl.append(player1_draw)
        player2_tbl.append(player2_draw)
        if i >= 5:
            at_war = False
```
